```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img=cv2.imread('mob.jpg')
img= cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

#crop image
# x,y,w,h=400,500,2000,2000
x,y,w,h=800,500,1200,2000
cr_img=img[y:y+h,x:x+w]

#affine transformation

rows,cols, _ = img.shape
print(rows,cols)

pt1=np.float32([[50,50],[200,50],[50,200]])
pt2=np.float32([[10,100],[200,50],[100,250]])

matrix=cv2.getAffineTransform(pt1,pt2)
af_img=cv2.warpAffine(img,matrix,(cols,rows))


#Brightness by log
#Converting to grayscale
gray_img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
#Applying log transformation
c=200/(np.log(1+np.max(gray_img)))
log_img=c*np.log(gray_img+1)
log_img=np.uint8(log_img)

#brightness by gamma
gamma=2.5
gamma_img=np.power(img/ 255.0,gamma)*255
gamma_img=np.uint8(gamma_img)
gamma_img_gray = cv2.cvtColor(gamma_img, cv2.COLOR_BGR2GRAY)
#negative
neg_img=255-img

plt.figure(figsize=(10,10))
plt.subplot(3,3,1)
plt.imshow(img)
plt.title('original')

plt.subplot(3,3,2)
plt.imshow(cr_img)
plt.axis = plt.gca().axis
plt.axis('off')
plt.title('cropped')

plt.subplot(3,3,3)
plt.imshow(af_img)
plt.axis = plt.gca().axis
plt.axis('off')
plt.title('affine')

plt.subplot(3,3,4)
plt.imshow(log_img, cmap='gray')
plt.axis = plt.gca().axis
plt.axis('off')
plt.title('log')

plt.subplot(3,3,5)
plt.imshow(gamma_img_gray, cmap='gray')
plt.axis = plt.gca().axis
plt.axis('off')
plt.title('gamma')

plt.subplot(3,3,6)
plt.imshow(neg_img,cmap='gray')
plt.axis = plt.gca().axis
plt.axis('off')
plt.title('negative')
plt.show()
```
